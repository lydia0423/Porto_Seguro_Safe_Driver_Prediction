### apply vs. transform

|                |                             agg                              |                            apply                             |                          transform                           |
| :------------: | :----------------------------------------------------------: | :----------------------------------------------------------: | :----------------------------------------------------------: |
|      描述      | <font color='red'>Aggregate</font> using <font color='#3198FB'>one or more operations</font> over the specified axis. | <font color='red'>Apply</font> <font color='#3198FB'>a function</font> along an axis of the DataFrame. | Call `func` on self producing a DataFrame with <font color='red'>the same axis</font> <font color='red'>shape as self</font>. |
|      函数      | Summary stats：mean、std<br />Numpy operations: np.mean/np.max<br />lambda（必须采用aggregate函数） |         Numpy operations: np.mean/np.max<br />lambda         | Summary stats：mean、std<br />Numpy operations: np.mean/np.max<br />lambda |
|      对象      |                             axis                             |                              df                              |                             axis                             |
| 不同列之间运算 |                             不能                             |                             可以                             |                             不能                             |
|   返回shape    |             和groupby的类别数一样<br />summarise             |             和groupby的类别数一样<br />summarise             |               和原数据行数一样<br />broadcast                |
